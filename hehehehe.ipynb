```python
def __init__(self, lr=0.001, n=1000):
        self.lr = lr
        self.n = n
        self.weights = None
        self.intercept = None

def fit(self, X, y):
        # Add a column of ones to X for the bias (intercept) term
        X = np.insert(X, 0, 1, axis=1)
        
        # Initialize weights and intercept
        self.weights = np.zeros(X.shape[1])
        
        # Gradient Descent for fitting the model
        for _ in range(self.n):
            # Calculate the predicted probabilities
            y_pred = self._sigmoid(np.dot(X, self.weights))
            
            # Update weights (including the intercept)
            gradient = np.dot((y_pred - y), X) / X.shape[0]
            self.weights -= self.lr * gradient

        # Split weights and intercept
        self.intercept = self.weights[0]
        self.weights = self.weights[1:]
        
def predict(self, X):
        # Add the bias term (column of ones) to X
        X = np.insert(X, 0, 1, axis=1)
        
        # Predict probabilities
        y_pred_prob = self._sigmoid(np.dot(X, np.insert(self.weights, 0, self.intercept)))
        
        # Convert probabilities to binary outcomes (0 or 1)
        return [1 if prob > 0.5 else 0 for prob in y_pred_prob]

def _sigmoid(self, X):
        # Sigmoid function to map predictions to probabilities
    return 1 / (1 + np.exp(-X))
```
